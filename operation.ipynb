```python
# Bitwise operations
a=80
b=100


print("a & b (AND):", a & b, a & b)
print("a | b (OR): ", a | b, (a | b))
print("a ^ b (XOR):", a ^ b, (a ^ b))
print("~a  (NOT): ", ~a,(~a))
print("a << 2 (Left Shift):", a << 2)
print("a >> 2 (Right Shift):", a >> 2)
```
