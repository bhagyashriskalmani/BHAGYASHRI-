```python
def load_data(filename):
    crops_data = []
    with open(filename, 'r') as f:
        for line in f:
            parts = line.strip().split(',')
            if len(parts) == 6:
                crop = {
                    "name": parts[0],
                    "season": parts[1],
                    "area": int(parts[2]),
                    "rainfall": int(parts[3]),
                    "fertilizer": int(parts[4]),
                    "yield": int(parts[5])
                }
                crops_data.append(crop)
    return crops_data

def display_crops(crops):
    print("\n--- Crop Data ---")
    for crop in crops:
        print(f"Name: {crop['name']}, Season: {crop['season']}, Area: {crop['area']}, Yield: {crop['yield']}, Rainfall: {crop['rainfall']}, Fertilizer: {crop['fertilizer']}")

def find_highest_and_lowest_yield(crops):
    if not crops:
        return None, None
    highest = max(crops, key=lambda crop: crop['yield'])
    lowest = min(crops, key=lambda crop: crop['yield'])
    return highest, lowest

def seasonal_summary(crops):
    summary = {}
    for crop in crops:
        season = crop['season']
        if season not in summary:
            summary[season] = {'total_yield': 0, 'count': 0}
        summary[season]['total_yield'] += crop['yield']
        summary[season]['count'] += 1
    for season, data in summary.items():
        summary[season]['average_yield'] = data['total_yield'] / data['count']
    return summary

def write_report(output_file, crops, highest, lowest, summary):
    with open(output_file, 'w') as f:
        f.write("Analysis Report\n")
        f.write("------------------\n\n")

        f.write("Crop Data:\n")
        for crop in crops:
            f.write(f"  Name: {crop['name']}, Season: {crop['season']}, Area: {crop['area']}, Yield: {crop['yield']}, Rainfall: {crop['rainfall']}, Fertilizer: {crop['fertilizer']}\n")
        f.write("\n")

        if highest and lowest:
            f.write(f"Highest Yield Crop: {highest['name']} with {highest['yield']} units\n")
            f.write(f"Lowest Yield Crop: {lowest['name']} with {lowest['yield']} units\n")
        f.write("\n")

        f.write("Seasonal Summary:\n")
        for season, data in summary.items():
            f.write(f"  {season}: Average Yield = {data['average_yield']:.2f}\n")

"""crop_count_per_season"""
def crop_count_per_season(crops):
    count = {}

    for crop in crops:
        season = crop["season"]
        if season not in count:
            count[season] = 0
        count[season] += 1

    for season, value in count.items():
        print(season, ":", value)
def fertilizer_efficiency(crop):
    return crop["yield"] / crop["fertilizer"]

def most_fertilizer_efficient_crop(crops):
    best = crops[0]

    for crop in crops:
        if fertilizer_efficiency(crop) > fertilizer_efficiency(best):
            best = crop

    print("Most Fertilizer Efficient Crop:", best["name"])
def rainfall_efficiency_analysis(crops, max_rainfall):
    print("Crops with rainfall <=", max_rainfall)

    for crop in crops:
        if crop["rainfall"] <= max_rainfall:
            print(crop["name"], "-", crop["yield"])

def find_max_rainfall(crops):
    max_rain = crops[0]["rainfall"]

    for crop in crops:
        if crop["rainfall"] > max_rain:
            max_rain = crop["rainfall"]

    return max_rain

def yield_per_area(crop):
    return crop["yield"] / crop["area"]

def rank_crops_by_yield(crops):
    ranked = sorted(crops, key=yield_per_area, reverse=True)

    rank = 1
    for crop in ranked:
        print(rank, crop["name"], "-", round(yield_per_area(crop), 2))
        rank += 1
def best_crop_per_season(crops):
    best = {}

    for crop in crops:
        season = crop["season"]

        if season not in best:
            best[season] = crop
        else:
            if yield_per_area(crop) > yield_per_area(best[season]):
                best[season] = crop

    for season, crop in best.items():
        print(season, ":", crop["name"])
def crops_in_multiple_seasons(crops):
    crop_seasons = {}

    for crop in crops:
        name = crop["name"]
        season = crop["season"]

        if name not in crop_seasons:
            crop_seasons[name] = []
        if season not in crop_seasons[name]:
            crop_seasons[name].append(season)

    for name, seasons in crop_seasons.items():
        if len(seasons) > 1:
            print(name, "grown in seasons:", seasons)
def recommend_crops(crops, season, max_rainfall, max_fertilizer):
    print("Recommended crops:")

    for crop in crops:
        if (crop["season"] == season and
            crop["rainfall"] <= max_rainfall and
            crop["fertilizer"] <= max_fertilizer):
            print(crop["name"])

def main():
  input_file="crops (1).txt"
  output_file="analysis-report.txt"

  crops=load_data(input_file)

  for crop in crops:
    print(crop)

  for crop in crops:
    for key,value in crop.items():
      print(f"{key}:{value}")
    print("----")

  display_crops(crops)

  highest,lowest=find_highest_and_lowest_yield(crops)

  summary=seasonal_summary(crops)

  write_report(output_file,crops,highest,lowest,summary)
  print("\n analysis complete.Report written to",output_file)
  crop_count_per_season(crops)
  most_fertilizer_efficient_crop(crops)
  max_rainfall = find_max_rainfall(crops)
  rainfall_efficiency_analysis(crops, max_rainfall)
  rank_crops_by_yield(crops)
  best_crop_per_season(crops)
  crops_in_multiple_seasons(crops)
  recommend_crops(crops, "summer", 100, 100)
main()
```
